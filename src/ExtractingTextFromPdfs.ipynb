````python
# Install image to PDF conversion
sudo apt install img2pdf
````

````python
# Installa OCR library tesseract and wrapper
sudo apt install ocrmypdf tesseract-ocr

# Install additional languages if needed
sudo apt install tesseract-ocr-deu tesseract-ocr-fra tesseract-ocr-ita
````

````python
# Convert PDF into images
#   * use 600dpi
#   * one PNG file per page
#   * output is named output-nnn.png, starting at output-001.png
pdftoppm -png -r 600 source.pdf output

# Convert images back into PDF
img2pdf *.png -o combined.pdf
````

````python
# Perform OCR and convert the combined PDF into a searchable PDF

ocrmypdf combined.pdf output_searchable.pdf

# For mximum compression
ocrmypdf \
  -l eng \
  --optimize 3 \
  --jpeg-quality 50 \
  --png-quality 50 \
  --output-type pdfa-2 \
  --oversample 300 \
  --skip-text \
  combined.pdf output_searchable_small.pdf

# For "balanced" compression
ocrmypdf \
  -l eng \
  --optimize 3 \
  --jpeg-quality 70 \
  --oversample 400 \
  --skip-text \
  combined.pdf output_searchable_balanced.pdf

# For "high quality" and less compression
ocrmypdf \
  -l eng \
  --optimize 2 \
  --jpeg-quality 80 \
  --oversample 400 \
  --skip-text \
  combined.pdf output_searchable_highq.pdf
````
