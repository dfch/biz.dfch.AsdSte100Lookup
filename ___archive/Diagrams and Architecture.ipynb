## Setup

We use Debian in a WSL2.

### Install Java through apt

```bash
sudo apt update
sudo apt install -y default-jre
```

### Find the Java version

```bash
java -version
```

The minimum result must be `17.0.17`.

```bash

```

### Install PlantUML through apt

```bash
sudo apt install plantuml -y
```

#### Alternative Install

```bash
mkdir -p ~/bin
cd ~/bin
wget https://github.com/plantuml/plantuml/releases/latest/download/plantuml.jar
java -jar ~/bin/plantuml.jar -version
```
